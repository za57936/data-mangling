```python
#storing data of users csv file into 'users' dataframe
users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv' )
users.head()
```

```python
#storing data of session csv file into 'session' dataframe
sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv' )
sessions.head()
```

```python
#storing data of products csv file into 'products' dataframe
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv' )
products.head()
```

```python
#storing data of transactions csv file into 'transactions' dataframe
transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv') 
transactions.head()
```

```python
#changing datatype of 'UserID' column in trasactions dataframe from float to int & filling zero in place of nan values
#to do further join operation on 'UserID' column datatype of users 'UserID' column(i.e. int) should match with it.
transactions['UserID'] = transactions['UserID'].fillna(0).astype(int)
transactions.head()
```

```python
#12 Joining users to transactions, keeping all rows from transactions & only matching rows from users (left join)
transactions.merge(users, on='UserID', how='left')
```

```python
transactions[~transactions['UserID'].isin(users['UserID'])]       #13 transactions having a UserID not in users
```

```python
#14 Joining users to transactions,keeping only rows from transactions & users that match via UserID (inner join)
pd.merge(transactions,users,how='inner',on='UserID')
```

```python
#15 Joining users to transactions,displaying all matching rows & all non-matching rows (full outer join)
transactions.merge(users, how='outer', on='UserID')
```

```python
#16 Sessions which occurred on same day each user registered
pd.merge(left=users, right=sessions, how='inner', left_on=['UserID', 'Registered'], right_on=['UserID', 'SessionDate'])
```

```python
#17 Building a dataset with every possible (UserID, ProductID) pair (cross join)
#storing  users 'UserID' in df1 dataframe
df1 = pd.DataFrame({'key': np.repeat(1, users.shape[0]), 'UserID': users.UserID})
#storing  products 'ProductID' in df2 dataframe
df2 = pd.DataFrame({'key': np.repeat(1, products.shape[0]), 'ProductID': products.ProductID})
#merging df1 & df1 based on 'key' 
pd.merge(df1, df2,on='key')[['UserID', 'ProductID']]
```

```python
#18 Determining how much quantity of each product was purchased by each user
#storing  users 'UserID' in df1 dataframe
df1 = pd.DataFrame({'key': np.repeat(1, users.shape[0]), 'UserID': users.UserID})
#storing  products 'ProductID' in df2 dataframe
df2 = pd.DataFrame({'key': np.repeat(1, products.shape[0]), 'ProductID': products.ProductID})
#merging df1 & df1 based on 'key' and storing it in 'user_products' dataframe
user_products = pd.merge(df1, df2,on='key')[['UserID', 'ProductID']]
#merging 'user_products' dataframe with 'transactions' dataframe on 'UserID & ProductID' 
#and after grouping them and summing the quantity applying the sum operation on quantity column
pd.merge(user_products, transactions, how='left', 
         on=['UserID', 'ProductID']).groupby(['UserID', 'ProductID']).apply(lambda x: pd.Series(dict(
         Quantity=x.Quantity.sum()))).reset_index().fillna(0)

```

```python
#19 For each user, getting each possible pair of pair transactions (TransactionID1,TransacationID2)
pd.merge(transactions, transactions, on='UserID')
```

```python
#20 Joining each user to his/her first occuring transaction in transactions table & storing in 'data' dataframe
pd.merge(users, transactions.groupby('UserID').first().reset_index(), how='left', on='UserID')
```

```python
#21 Testing to see if we can drop columns
data = pd.merge(users, transactions.groupby('UserID').first().reset_index(), how='left', on='UserID')
my_columns = list(data.columns)     #getting all columns of 'data' dataframe in my_columns
my_columns
```

```python
data.dropna().shape     # drop columns for even a single nan
```

```python
data.dropna(how='all', axis=1).shape       # drop columns for all nan

```

```python
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns)  #setting threshold

```

```python
missing_info = list(data.columns[data.isnull().any()])    #missing info
missing_info
```

```python
for col in missing_info:       #Count of missing data
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
for col in missing_info:       #percentage of missing data
    percent_missing = data[data[col].isnull() == True].shape[0] /data.shape[0]
    print('percent missing for column {}: {}'.format( col,percent_missing))
```
